## Feature correlations

```python
metric = "F1" # Choices: "Accuracy", "F1", "Log Loss"
```

```python
tuned_alg_perf = get_tuned_alg_perf(metadataset_df, metric=metric)
tuned_alg_perf
```

```python
joined_df = tuned_alg_perf.merge(metafeatures_df, right_on="dataset_name", left_on="dataset_fold_id", how='left')
```

```python
feature_corrs = compute_feature_corrs(joined_df, metric)
```

```python
feature_corrs
```

```python
def top_k_mfs_corr(algos, k):

    features = ["feature_"+str(i+1) for i in range(k)]
    df = pd.DataFrame(columns = features)
    pd.set_option('max_colwidth', 400)
    
    for algo in algos:
        corr_features = feature_corrs.reindex(feature_corrs[algo].abs().sort_values(ascending=False).index)
        features = corr_features.index[:k].tolist()
        scores = corr_features[algo].tolist()[:k]
        feature_scores = [f"{features[i]}, score:{scores[i]}" for i in range(k)]
        df.loc[algo] = feature_scores
        
    return df
```

```python
top_k_mfs_corr(metadataset_df['alg_name'].unique(), 9)
```
