## Correlations between pairs of algorithms

```python
def alg_performance_difference(tuned_alg_perf, metric, alg_pair):
    """
    Get the difference in performance between the algorithms specified in alg_pair for the chosen metric.
    Args:
        tuned_alg_perf: tuned algorithm performance, as output by get_tuned_alg_perf()
        metric: the name of the metric
        alg_pair: tuple or list with two strings representing the names of the two algorithms. The difference is
            computed as alg_pair[0] - alg_pair[1]
    """
    perf_pair = [tuned_alg_perf[tuned_alg_perf["alg_name"] == alg].set_index("dataset_fold_id") for alg in alg_pair]
    delta = perf_pair[0][f"{metric}__test"] - perf_pair[1][f"{metric}__test"]
    delta = delta.loc[perf_pair[0].index.intersection(perf_pair[1].index)]
    return delta
```

```python
def alg_performance_matrix(tuned_alg_perf, metric):
    """
    Compute the pairwise performance differences between all algorithms. Returns both the average performance difference
    per pair (mean_deltas) and all of the performance differences (all_deltas)

    Args:
        tuned_alg_perf: tuned algorithm performance, as output by get_tuned_alg_perf()
        metric: name of the metric
    """
    all_deltas = []
    mean_deltas = []
    for alg_pair in itertools.combinations(tuned_alg_perf["alg_name"].unique(), 2):
        delta = alg_performance_difference(tuned_alg_perf, metric, alg_pair)
        delta.name = "{}-{}".format(*alg_pair)
        all_deltas.append(delta)
        mean_deltas += [
            {"pair": alg_pair, "mean_delta": delta.mean()},
            {"pair": (alg_pair[1], alg_pair[0]), "mean_delta": -delta.mean()},
        ]
    all_deltas = pd.concat(all_deltas, axis=1).T

    mean_deltas = pd.DataFrame(mean_deltas).set_index("pair")
    mean_deltas.index = pd.MultiIndex.from_tuples(mean_deltas.index, names=["Algorithm 1", "Algorithm 2"])
    
    return mean_deltas, all_deltas
```

```python
def alg_pair_corrs(tuned_alg_perf, metafeatures_df, metric, alg_pair, as_abs=False):
    """
    Computes all correlations between all metafeatures and the performance difference between algorithms 1 and 2.

    Args:
        tuned_alg_perf: tuned algorithm performance, as output by get_tuned_alg_perf()
        metafeatures_df: dataframe of metafeatures, as output by process_metafeatures()
        metric: name of the metric
        alg_pair: tuple or list with names of two algorithms
        as_abs: if True, returns absolute correlations. Defaults to False.
    """
    delta = alg_performance_difference(tuned_alg_perf, metric, alg_pair)
    delta.name = "delta_perf"

    delta_mf = metafeatures_df.merge(delta, left_on="dataset_name", right_index=True, how='right')
    corr_columns = [col_name for col_name in metafeatures_df if col_name.startswith("f__")]
    results = delta_mf[corr_columns].corrwith(delta_mf["delta_perf"])
    if as_abs:
        return results.abs().sort_values(ascending=False)
    else:
        return results.sort_values(ascending=False)
```

```python
def all_alg_pair_corrs(tuned_alg_perf, metafeatures_df, metric):
    """
    Computes all correlations between all metafeatures and the performance difference of each pair of algorithms.
    Returns a dataframe with all max correlations per algorithm pair (max_corrs) and one with all of the correlations 
    for all algorithm pairs (all_pair_corrs)
    
    Args:
        tuned_alg_perf: tuned algorithm performance, as output by get_tuned_alg_perf()
        metafeatures_df: dataframe of metafeatures, as output by process_metafeatures()
        metric: name of the metric

    """
    all_pair_corrs = []

    max_corrs = []
    for alg_pair in itertools.combinations(tuned_alg_perf["alg_name"].unique(), 2):
        pair_corr = alg_pair_corrs(tuned_alg_perf, metafeatures_df, metric, alg_pair)
        pair_corr.name = "{}-{}".format(*alg_pair)
        all_pair_corrs.append(pair_corr)
        idxmax = pair_corr.abs().idxmax()
        corr = pair_corr[idxmax]
        max_corrs += [
            {"pair": alg_pair, "max_corr": corr, "max_abs": abs(corr), "metafeature": idxmax},
            {"pair": (alg_pair[1], alg_pair[0]), "max_corr": corr, "max_abs": abs(corr), "metafeature": idxmax},
            #{"pair": (alg_pair[1], alg_pair[0]), "max": -pair_corr.min(), "idxmax": pair_corr.idxmin()},
        ]
    all_pair_corrs = pd.concat(all_pair_corrs, axis=1).T

    max_corrs = pd.DataFrame(max_corrs).set_index("pair")
    max_corrs.index = pd.MultiIndex.from_tuples(max_corrs.index, names=["Algorithm 1", "Algorithm 2"])
    
    return max_corrs, all_pair_corrs
```

```python
def plot_matrix(matrix, cmap='viridis', title=None):
    cellsize = 5
    fig, ax = plt.subplots(1,1,figsize=(cellsize*1.2, cellsize*matrix.shape[0]/matrix.shape[1]))
    im = ax.pcolormesh(matrix, edgecolor='k', linewidth=0.5, cmap=cmap) 
    if title is not None:
        plt.title(title)
    plt.gca().invert_yaxis()
    ax.set_xticks([])
    ax.set_xticklabels('')
    ax.set_xticks([c+0.5 for c in range(matrix.shape[0]-0)], minor=True)

    ax.set_xticklabels(matrix.columns, minor=True, rotation=90)
    ax.set_yticks([])
    ax.set_yticklabels('')    
    ax.set_yticks([c+0.5 for c in range(matrix.shape[1]-0)], minor=True)
    ax.set_yticklabels(matrix.index, minor=True)
    fig.colorbar(im, ax=ax)
        
```

```python
metrics = ["F1", "Log Loss", "Accuracy"]
```

```python
max_corrs, all_pair_corrs = {}, {}
mean_deltas, all_deltas = {}, {}
for metric in metrics:
    tuned_alg_perf = get_tuned_alg_perf(metadataset_df, metric=metric)
    joined_df = tuned_alg_perf.merge(metafeatures_df, right_on="dataset_name", left_on="dataset_fold_id", how='left')
    
    max_corrs[metric], all_pair_corrs[metric] = all_alg_pair_corrs(tuned_alg_perf, metafeatures_df, metric)
    mean_deltas[metric], all_deltas[metric] = alg_performance_matrix(tuned_alg_perf, metric)
```

```python
def show_metric_corr_results(metric):
    """
    Display a plot of max correlation per algorithm pair, mean difference per algorithm pair, and display the table
    corresponding to the names of the metafeatures of the first plot.
    
    Args:
        metric: name of the metric
    """
    comp_matrix = max_corrs[metric]["max_abs"].unstack()
    plot_matrix(comp_matrix, title=f"Max Correlation with Performance Difference ({metric})")
    plot_matrix(mean_deltas[metric]["mean_delta"].unstack(), title=f"Performance Difference Mean ({metric})", cmap="RdBu")
    feat_names = max_corrs[metric]["metafeature"].unstack()
    feat_names = feat_names.applymap(lambda entry: entry.replace("f__pymfe.", "") if type(entry) is str else entry)
    display_html(feat_names)
```

```python
def stat_summary(metric, k=10):
    """
    Display tables summarizing the algorithm performance differences, the max correlations between metafeatures and 
    algorithm performance difference, and the top values for each of these.
    
    Args:
        metric: name of the metric
    """
    
    all_pairs = list(itertools.combinations(tuned_alg_perf["alg_name"].unique(), 2))
    float_format = '%.3f'
    
    pretty_corr_table = pd.DataFrame(index=max_corrs[metric].loc[all_pairs].index)
    pretty_corr_table["Abs. Correlation"] = max_corrs[metric].loc[pretty_corr_table.index]["max_abs"]
    pretty_corr_table["Metafeature"] = max_corrs[metric].loc[pretty_corr_table.index]["metafeature"].apply(lambda entry: entry.replace("f__pymfe.", "") if type(entry) is str else entry)

    pretty_diff_table = mean_deltas[metric][mean_deltas[metric]["mean_delta"] > 0]
    pretty_diff_table.columns = ["Mean Difference"]

    top_k_corrs = pretty_corr_table.sort_values(by="Abs. Correlation", ascending=False).iloc[:k]
    top_k_diffs = pretty_diff_table.loc[pretty_diff_table.abs().sort_values( by="Mean Difference", ascending=False).index].iloc[:k]


    top_k_corrs.to_csv(f"TopCorrs_{metric}.csv", float_format=float_format)
    top_k_diffs.to_csv(f"TopDiffs_{metric}.csv", float_format=float_format)
    
    pretty_corr_table.to_csv(f"AllCorrs_{metric}.csv", float_format=float_format)
    pretty_diff_table.to_csv(f"AllDiffs_{metric}.csv", float_format=float_format)

    display_html(top_k_corrs)
    display_html(top_k_diffs)

    display_html(pretty_corr_table)
    display_html(pretty_diff_table)
```
